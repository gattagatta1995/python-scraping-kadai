```python
import requests
import pprint
from getpass import getpass

# APIキーの設定
api_key = getpass('取得したAPIキーをここに貼り付けてください')

# 検索対象の駅を変数addressに格納する
address = input('検索する駅名を入力して下さい。')

# 入力された駅の緯度経度を取得する
url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データの取得
response = requests.get(
    url,
    params={
        'key': api_key,
        'address': address + '駅' # Changed station_name to address
    }
)

# レスポンスのJSONを取得
json_data = response.json()


# 緯度・経度を取得
location = json_data['results'][0]['geometry']['location']
latitude = location['lat']
longitude = location['lng']

print(f'\n【検索駅】{address}')
print(f'緯度：{latitude}, 経度：{longitude}\n')

# レストラン情報を取得するAPI
restaurant_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

# レストランデータの取得
restaurant_response = requests.get(
    restaurant_url,
    params={
        'location': f'{latitude},{longitude}',
        'radius': 500,
        'type': 'restaurant',
        'language': 'ja',
        'key': api_key
    }
)

restaurant_data = restaurant_response.json()

# レストラン情報の出力
print('【半径500m以内のレストラン一覧】\n')

for restaurant in restaurant_data.get('results', []):
    name = restaurant.get('name', '情報なし')
    rating = restaurant.get('rating', '評価なし')
    address = restaurant.get('vicinity', '住所なし')

    print('------------------------------')
    print(f'店名　：{name}')
    print(f'評価　：{rating}')
    print(f'住所　：{address}')

```
